```python
## 변경할 파일이 위치한 폴더
dir_path = r'C:\Users\Hojun\Desktop\academic_project\labels(000000~002056)'
mistake_file = [] # 오분류된 파일들

# file_names에서 txt_file 뽑기 
for txt_name in file_names:
    filepath = os.path.join(dir_path, txt_name)

    # 각 파일 열기
    with open(filepath, 'r') as f:
        lines = f.readlines()
        detect_count = len(lines) ## label 갯수
        # label 갯수
        zero_cnt = 0 
        for i in range(detect_count):
            # i번째 label 확인
            if lines[i][0] == '0':
                # label 3
                lines[i] = '3' + lines[i][1:]
                
                # 파일 수정 - label 3 변경
                with open(filepath, 'w') as f:
                    f.writelines(lines)
                zero_cnt +=1

        # 파일 내에 0이 없다면 오분류한 파일
        if zero_cnt == 0:
            mistake_file.append(txt_name)

print(mistake_file)
```

```python
dir_path = r"C:\Users\Hojun\Desktop\academic_project\labels(000000~002056)"
#dir_path = r"C:\Users\Hojun\Desktop\academic_project\multi"

file_names = []
for folder, subfolders, filenames in os.walk(dir_path): ## 폴더 내 모든 파일 탐색
    for img in filenames:
        img = img.split('.')[0] + '.txt' ## jpg 확장자 없애고 txt로 변환
        file_names.append(img)
print(file_names)
```

```python
car = 0
motor =0
person =0
trailer =0

# file_names에서 txt_file 뽑기 
for txt_name in file_names:
    filepath = os.path.join(dir_path, txt_name)
    # 각 파일 열기
    with open(filepath, 'r') as f:
        lines = f.readlines()
        detect_count = len(lines) ## label 갯수
        # label 갯수
        label = []
        for i in range(detect_count):
            label.append(lines[i][0])
            
        if '0' in label:
            car +=1
        if '1' in label:
            motor +=1
        if '2' in label:
            person +=1
        if '3' in label:
            trailer +=1
               
print(f'car : {car}')
print(f'motor : {motor}')
print(f'person : {person}')
print(f'trailer : {trailer}')

```

```python
import os

folder_path = "/path/to/your/image/folder"  # Replace with your folder path
prefix = "new_"  # Change this to the desired prefix

# List all files in the folder
file_names = os.listdir(folder_path)

# Iterate over the files
for file_name in file_names:
    # Check if the file is an image (by checking the file extension)
    if file_name.lower().endswith((".png", ".jpg", ".jpeg", ".gif", ".bmp", ".tiff")):
        # Create the new file name with the prefix
        new_file_name = prefix + file_name

        # Create the full paths for the original and new file names
        old_file_path = os.path.join(folder_path, file_name)
        new_file_path = os.path.join(folder_path, new_file_name)

        # Rename the file
        os.rename(old_file_path, new_file_path)

print("Image file names updated.")
```
