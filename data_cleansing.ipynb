```python
import os
import shutil
import numpy as np
from sklearn.model_selection import train_test_split

source_image_dir = r"C:\Users\Hojun\Desktop\academic_project\Images"
source_label_dir = r"C:\Users\Hojun\Desktop\academic_project\Labels"

train_image_dir = r"C:\Users\Hojun\Desktop\academic_project\split\train\images"
train_label_dir = r"C:\Users\Hojun\Desktop\academic_project\split\train\labels"

val_image_dir = r"C:\Users\Hojun\Desktop\academic_project\split\valid\images"
val_label_dir = r"C:\Users\Hojun\Desktop\academic_project\split\valid\labels"

test_image_dir = r"C:\Users\Hojun\Desktop\academic_project\split\test\images"
test_label_dir = r"C:\Users\Hojun\Desktop\academic_project\split\test\labels"

# Create the destination folders if they don't exist
os.makedirs(train_image_dir, exist_ok=True)
os.makedirs(train_label_dir, exist_ok=True)
os.makedirs(val_image_dir, exist_ok=True)
os.makedirs(val_label_dir, exist_ok=True)
os.makedirs(test_image_dir, exist_ok=True)
os.makedirs(test_label_dir, exist_ok=True)

# List all image files in the source directory
image_files = [f for f in os.listdir(source_image_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Split file list into train and test sets (80% train, 20% test)
train_images, test_images = train_test_split(image_files, test_size=0.1, shuffle=True, random_state=42)

# Further split the train files into train and validation sets (80% train, 20% validation)
train_images, val_images = train_test_split(train_images, test_size=1/9, shuffle=True, random_state=42)

# Function to copy both image and label files
def copy_files(file_list, source_image_dir, source_label_dir, dest_image_dir, dest_label_dir):
    for file in file_list:
        # Copy image file
        shutil.copy(os.path.join(source_image_dir, file), os.path.join(dest_image_dir, file))
        
        # Get the corresponding label file
        label_file = os.path.splitext(file)[0] + '.txt'
        
        # Copy label file
        shutil.copy(os.path.join(source_label_dir, label_file), os.path.join(dest_label_dir, label_file))

# Move the files to their respective folders
copy_files(train_images, source_image_dir, source_label_dir, train_image_dir, train_label_dir)
copy_files(val_images, source_image_dir, source_label_dir, val_image_dir, val_label_dir)
copy_files(test_images, source_image_dir, source_label_dir, test_image_dir, test_label_dir)
```
